```python
from sympy import init_printing, Matrix, symbols, cos, sin ,sqrt, pi, I
init_printing()
from opg33 import create_matrix
import numpy as np
```

```python
## TESTDATA
a, N = 1, 10
rho ,phi = np.zeros(N),np.zeros(N)
for i in range(N):
    rho[i] = sqrt(2)*a*i/N-sqrt(2)*a/2
    phi[i] = i*pi/N
rho, phi = np.array(rho), np.array(phi)
A = create_matrix(a,N,rho,phi)
A_round = np.zeros(A.shape)
for i, row in enumerate(A):
    for j, element in enumerate(row):
        A_round[i,j] = np.round(element,8)
A, A_round = Matrix(A), Matrix(A_round)
A_round.T
```

```python
import pickle
ATA = A.T*A
eigens = list(ATA.eigenvals().keys())
real_eigens = [eig for eig in eigens if eig.is_real]

with open('egenværdier/eigs.txt','w') as f:
    for eig in eigens:
        f.write(str(eig)+'\n')
with open('egenværdier/eigs_pickle.pkl','wb') as f:
    pickle.dump(eigens,f)

with open('egenværdier/real_eigs.txt','w') as f:
    for eig in real_eigens:
        f.write(str(eig)+'\n')
with open('egenværdier/real_eigs_pickle.pkl','wb') as f:
    pickle.dump(real_eigens,f)
```
